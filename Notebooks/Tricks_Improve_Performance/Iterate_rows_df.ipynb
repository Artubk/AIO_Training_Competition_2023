# Iterrows

```python
def iterrows(df):
    list_id = []
    for index, row in df.iterrows():
        list_id.append(row["id"])

    return list_id
```

```python
list_id = iterrows(df)
```

```python
len(list_id)
```

# Optimize

```python
def np_vectorization(df):
    np_arr = df.to_numpy()

    return np_arr[:,0]
```

```python
list_id = np_vectorization(df)
```
